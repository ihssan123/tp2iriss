```python
import csv

def process_word(word, pro, en):
    result = []
    i = 0
    word2 = word[i:len(word)]
    result.append(word2)
    while i < len(word):
        wordd = word[0:i]
        if wordd in pro:
            word2 = word[i:len(word)]
            result.append(word2)

            j = len(word)
            while j > 0:
                wordd3 = word[j:len(word)]
                if wordd3 in en:
                    word22 = word[0:j]
                    word222 = word[i:j]
                    result.append(word22)
                    result.append(word222)
                j = j - 1
        i += 1
    return result

def remove_duplicates(word_list):
    unique_words = set(word_list)
    result = list(unique_words)
    return result

def mots_meme_longueur(mot, tableau_mots):
    longueur_mot = len(mot)
    mots_de_meme_longueur = [mot for mot in tableau_mots if len(mot) == longueur_mot]
    return mots_de_meme_longueur

def recherche_caractere(tableau_mots, motttt):
    caracteres_inattendus = []
    mots_sans_caractere_inattendu = []
    mots_sans_caractere_attendu = []

    for mot in tableau_mots:
        caractere_trouve = False
        for position, caractere in enumerate(mot):
            if caractere not in ['ف', 'ع', 'ل']:
                if position < len(motttt) and caractere != motttt[position]:
                    caracteres_inattendus.append((caractere, position))
                    if mot not in mots_sans_caractere_inattendu:
                        mots_sans_caractere_inattendu.append(mot)
                    caractere_trouve = True
                    break

        if not caractere_trouve:
            mots_sans_caractere_attendu.append(mot)

    return caracteres_inattendus, mots_sans_caractere_inattendu, mots_sans_caractere_attendu

def construire_mot(mot1, mot2):
    if len(mot1) != len(mot2):
        return "Les mots n'ont pas la même taille."

    nouveau_mot = ""
    for i, caractere in enumerate(mot1):
        if caractere in ['ف', 'ع', 'ل']:
            nouveau_mot += mot2[i]

    return nouveau_mot

# utilisation :

# Définir le nom des fichiers CSV
nom_fichier_enclitiques = "/content/files/liste_enclitiques (2).csv"
nom_fichier_proclitiques = "/content/files/liste_proclitiques (2).csv"

# Créer des listes vides pour stocker les données
pro = []
en = []

# Lire le fichier CSV pour les proclitiques
with open(nom_fichier_proclitiques, newline='', encoding='utf-8') as csvfile:
    lecteur = csv.reader(csvfile, delimiter=',')
    for row in lecteur:
        for item in row:
            pro.append(item)

# Lire le fichier CSV pour les enclitiques
with open(nom_fichier_enclitiques, newline='', encoding='utf-8') as csvfile:
    lecteur = csv.reader(csvfile, delimiter=',')
    for row in lecteur:
        for item in row:
            en.append(item)

mot = input("Entrez un mot : ")
resultats = process_word(mot, pro, en)
print("Combinations possibles de proclitiques et enclitiques jusqu'à obtenir le mot complet :")
print(resultats)

```

```python
def find_matching_schemas(word_list, schema_list):
    matching_words_with_schemas = {}
    for word in word_list:
        matching_schemas = [schema for schema in schema_list if len(schema) == len(word)]
        if matching_schemas:
            matching_words_with_schemas[word] = matching_schemas

    return matching_words_with_schemas

# Example usage:

# Your word list
word_list = ['كتابتهن', 'تابتهن', 'كتابت', 'تابت']

# Your schema list


# Finding words and their matching schemas
results = find_matching_schemas(word_list, values_list)

# Displaying the results
for word, schemas in results.items():
    print(f"Word: {word}, Matching Schemas: {schemas}")
```

```python
def find_closest_schema(word_list, schema_list):
    closest_schemas = {}
    for word in word_list:
        closest_distance = float('inf')  # Initialize with infinity
        closest_schema = None
        for schema in schema_list:
            if len(schema) == len(word):
                dist = distance(schema, word)
                if dist < closest_distance:
                    closest_distance = dist
                    closest_schema = schema

        if closest_schema is not None:
            closest_schemas[word] = closest_schema

    return closest_schemas
```

```python
def racine(mot, sheme):
    f, a, l = None, None, None  # Initialisez les variables à None pour le cas où les caractères ne sont pas trouvés
    for index, caract in enumerate(sheme):
        if caract in ['ف', 'ع', 'ل']:
            if caract == 'ف':
                f = index
            elif caract == 'ع':
                a = index
            elif caract == 'ل':
                l = index

    # Vérifiez si tous les indices ont été trouvés avant d'accéder aux éléments du mot
    if f is not None and a is not None and l is not None:
        ff = mot[f]
        aa = mot[a]
        ll = mot[l]
        return ff, aa, ll
    else:
        return None  # Retournez quelque chose pour indiquer qu'au moins un caractère n'a pas été trouvé

#exemple
mot = "داهم"
sheme = "فاعل"

resultat = racine(mot, sheme)

if resultat:
    ff, aa, ll = resultat
    print(f"La racine du mot {mot} avec le schème {sheme} est : {ff} {aa} {ll}")
else:
    print(f"Impossible de trouver la racine du mot {mot} avec le schème {sheme}")
```
